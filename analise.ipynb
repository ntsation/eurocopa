```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('euro2024_players.csv')
df
```

```python
idade_distribuicao = df['Age'].value_counts().sort_index()
# Plotando a distribuição de idade
idade_distribuicao.plot(kind='bar', title='Distribuição de Idade')
plt.xlabel('Idade')
plt.ylabel('Número de Jogadores')
plt.show()
```

```python
# Média de altura por posição
altura_por_posicao = df.groupby('Position')['Height'].mean()
print(altura_por_posicao)

# Plotando a média de altura por posição
altura_por_posicao.plot(kind='bar', title='Altura Média por Posição')
plt.xlabel('Posição')
plt.ylabel('Altura Média (cm)')
plt.show()
```

```python
# Contagem de jogadores por pé dominante
pe_dominante = df['Foot'].value_counts()
print(pe_dominante)

# Plotando a contagem de jogadores por pé dominante
pe_dominante.plot(kind='bar', title='Pé Dominante dos Jogadores')
plt.xlabel('Pé Dominante')
plt.ylabel('Número de Jogadores')
plt.show()
```

```python
# Média de gols por idade
gols_por_idade = df.groupby('Age')['Goals'].mean()
print(gols_por_idade)

# Plotando a média de gols por idade
gols_por_idade.plot(kind='line', title='Média de Gols por Idade')
plt.xlabel('Idade')
plt.ylabel('Média de Gols')
plt.show()
```

```python
# Valor de mercado médio por posição
valor_mercado_por_posicao = df.groupby('Position')['MarketValue'].mean()
print(valor_mercado_por_posicao)

# Plotando o valor de mercado médio por posição
valor_mercado_por_posicao.plot(kind='bar', title='Valor de Mercado Médio por Posição')
plt.xlabel('Posição')
plt.ylabel('Valor de Mercado Médio (€)')
plt.show()
```

```python
# Contagem de jogadores por clube
jogadores_por_clube = df['Club'].value_counts()
print(jogadores_por_clube)

# Plotando a contagem de jogadores por clube
jogadores_por_clube.plot(kind='bar', title='Número de Jogadores por Clube', figsize=(10, 6))
plt.xlabel('Clube')
plt.ylabel('Número de Jogadores')
plt.show()
```

```python
# Contagem de jogadores por país
jogadores_por_pais = df['Country'].value_counts()
print(jogadores_por_pais)

# Plotando a contagem de jogadores por país
jogadores_por_pais.plot(kind='bar', title='Número de Jogadores por País', figsize=(10, 6))
plt.xlabel('País')
plt.ylabel('Número de Jogadores')
plt.show()
```

```python
# Filtrar jogadores com idade de 35 anos ou mais
jogadores_35_ou_mais = df[df['Age'] >= 35]

# Ordenar por gols em ordem decrescente
jogadores_mais_gols = jogadores_35_ou_mais.sort_values(by='Goals', ascending=False)
print(jogadores_mais_gols.head(5))
```
